```python
import numpy as np

def calculate_majority(row):
    numerical_values = [row['Sentiment_Numerical_1'], row['Sentiment_Numerical_2'], row['Sentiment_Numerical_3']]
    positive_count = numerical_values.count(1)
    negative_count = numerical_values.count(-1)
    neutral_count = numerical_values.count(0)
    
    if positive_count > negative_count and positive_count > neutral_count:
        return 1
    elif negative_count > positive_count and negative_count > neutral_count:
        return -1
    else:
        return 0

df1["Sentiment"] = df1.apply(calculate_majority, axis=1)
```
