```python
import pandas as pd
from fuzzywuzzy import fuzz, process
sentiment_df = pd.read_csv("annotation.csv")

comment_id_df =pd.read_csv("EDA.csv")
# Iterate through each sentiment comment and find the closest matching comment ID

comment_id_df
```

```python
sentiment_df
```

```python
comment_id_mapping = {}
for sentiment_comment in sentiment_df["Comments"]:
    closest_match = process.extractOne(sentiment_comment, comment_id_df["Comment Text"])
    comment_id_mapping[sentiment_comment] = closest_match[2]

# Add the matched Comment IDs to the sentiment data frame
sentiment_df["Comment ID"] = sentiment_df["Comments"].map(comment_id_mapping)
```
