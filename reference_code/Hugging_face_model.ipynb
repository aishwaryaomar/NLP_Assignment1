```python
from transformers import pipeline
import pandas as pd
from collections import Counter

# Load the sentiment analysis pipelines with the specified models
sentiment_task3 = pipeline("sentiment-analysis", model="Seethal/sentiment_analysis_generic_dataset")

# Load the DataFrame containing comments
df_comments = pd.read_csv("reddit_comments.csv")  # Replace with your file path

# Analyze each comment and get sentiment labels using both models
sentiments3 = []

for comment in df_comments["Comments"]:
    # Truncate the comment to fit within the model's maximum sequence length
    truncated_comment = comment[:128]

    result3 = sentiment_task3(truncated_comment)
    s3 = result3[0]["label"]
    sentiments3.append(s3)

# Add sentiment columns to the DataFrame
df_comments["Sentiment_3"] = sentiments3

# Display the DataFrame with sentiment analysis results
print(df_comments)
```

```python
import pandas as pd
from collections import Counter
```

```python
# Combine the predictions into a list of DataFrames
all_predictions = [predictions_model1, predictions_model2, predictions_model3]

# Combine predictions using majority vote
majority_labels = []
for comments in zip(*[df['Comments'] for df in all_predictions]):
    label_counts = Counter([df[df['Comments'] == comment]['Sentiment_Numerical'].values[0] for df, comment in zip(all_predictions, comments)])

#    label_counts = Counter([df[df['Comments'] == comment]['Sentiment_Numerical'].values[0] for df, comment in zip(all_predictions, comments)])
    majority_label = label_counts.most_common(1)[0][0]
    majority_labels.append(majority_label)
```

```python
predictions_model1['majority_label'] = majority_labels
```

```python
from sklearn.datasets import make_classification
from sklearn.preprocessing import LabelEncoder
from imblearn.over_sampling import SMOTE
```

```python
df = pd.DataFrame(predictions_model1)
```

```python
# Convert string categories to numerical values using LabelEncoder
label_encoder = LabelEncoder()
df['x_encoded'] = label_encoder.fit_transform(df['Comments'])

# Separate features (X) and target (y)
X = df[['x_encoded']]
y = df['majority_label']

# Apply SMOTE to balance the dataset
smote = SMOTE()
X_resampled, y_resampled = smote.fit_resample(X, y)

# Convert numerical categories back to original string categories
resampled_df = pd.DataFrame({'x': label_encoder.inverse_transform(X_resampled), 'y': y_resampled})
```
